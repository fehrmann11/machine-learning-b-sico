```python
#importa numpy y pandas
import numpy as np
import pandas as pd
```

```python
#200 datos, con 50 filas y 4 columnas
df = pd.DataFrame(np.random.randint(200,size=(50,4)),columns=['a','b','c','d']) 
df
```

```python
df['a'].hist()
```

```python
df['b'].hist(bins=30) #30 FILAS
```

```python
df['c'].plot.hist() #plot es más lindo
```

```python
df.plot.area(alpha=0.3) #area de los valores de las columnas a b c y d
```

```python
#gráfico de barras
df.plot.bar(stacked = True) #acumula los valores 
```

```python
df.plot.scatter(x='a',y='b') #hace puntos entre 0 y 200 con eje x e y correspondiente a las columnas a y b
```

```python
df.plot.scatter(x='a',y='b',c='c', cmap='coolwarm') #3 dimensiones
```

```python
#grafico de caja
df.plot.box()
```

```python
#grafico hexadecimal
df.plot.hexbin(x='a',y='b',gridsize=12)
```

```python
#gráfico kde
df.plot.kde()
```

```python
#gráfico kde
df.plot.density()
```
